```python
# Combine Category 1, 2, and 3 into a single textual input
df['combined_categories'] = df['Category 1'].astype(str) + ' ' + df['Category 2'].astype(str) + ' ' + df['Category 3'].astype(str)
```

```python
# Tokenize the combined categories into words
tokenized_categories = df['combined_categories'].apply(lambda x: x.split())

# Train Word2Vec on the tokenized data
w2v_model = Word2Vec(sentences=tokenized_categories, vector_size=50, window=5, min_count=1, workers=4)

# Create embeddings for each row by averaging word embeddings
def get_embedding(text):
    tokens = text.split()
    vectors = [w2v_model.wv[word] for word in tokens if word in w2v_model.wv]
    if len(vectors) > 0:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(50)  # Return zero vector if no words match

# Apply embeddings to the dataset
df['category_embedding'] = df['combined_categories'].apply(get_embedding)
```

```python
# Stack all embeddings into a matrix
embeddings = np.stack(df['category_embedding'].values)

# Cluster into 50 groups
kmeans = KMeans(n_clusters=50, random_state=42)
df['Category_Group'] = kmeans.fit_predict(embeddings)
```

```python
from collections import Counter
df['Cluster'] = kmeans.labels_
# Analyze top words in each cluster
for cluster in range(kmeans.n_clusters):
    print(f"Cluster {cluster}:")
    cluster_data = df[df['Cluster'] == cluster]['combined_categories']
    word_counter = Counter(" ".join(cluster_data).split())
    common_words = word_counter.most_common(10)  # Get top 10 words
    print("Top words:", common_words)
    print("\n")
```

## Preprocessing

```python
df= pd.read_csv('ds_word2vec.csv')


#df['Category 1'] = df['Category 1'].astype('category')
#df['Category 2'] = df['Category 2'].astype('category')
df['Category'] = df['Category'].astype('category')
df['Client_Country'] = df['Client_Country'].astype('category')


# Create target variables based on Payment_type_Hourly
df['Hourly_Cost'] = np.where(df['Payment_type_Hourly'] == 1, df['Hourly_Rate'], np.nan)
df['Fixed_Cost'] = np.where(df['Payment_type_Hourly'] == 0, df['Job_Cost'], np.nan)
```

```python
# Compute group statistics
group_stats = df.groupby(['Client_Country', 'Category'])['Fixed_Cost'].agg(['mean', 'std']).reset_index()

# Merge the stats back to the dataframe
df = df.merge(group_stats, on=['Client_Country', 'Category'], how='left')

# Function to replace outliers with group median
def replace_outliers_with_median(row):
    median = row['mean']  # Use median instead of mean
    std = row['std']
    lower_bound = median - 2 * std
    upper_bound = median + 2 * std
    if row['Fixed_Cost'] < lower_bound or row['Fixed_Cost'] > upper_bound:
        return median  # Replace outliers with group median
    return row['Fixed_Cost']  # Keep original value if not an outlier

# Apply the function to adjust the outliers
df['Fixed_Cost'] = df.apply(replace_outliers_with_median, axis=1)

# Drop unnecessary columns
df = df.drop(['mean', 'std'], axis=1)

# Verify the adjusted dataset
print(df[['Client_Country', 'EX_level_demand', 'Category', 'Fixed_Cost']].head())
```

```python
# Compute group statistics
group_stats = df.groupby(['Client_Country','Category'])['Hourly_Cost'].agg(['mean', 'std']).reset_index()

# Merge the stats back to the dataframe
df = df.merge(group_stats, on=['Client_Country','Category'], how='left')

# Function to replace outliers with group median
def replace_outliers_with_median(row):
    median = row['mean']
    std = row['std']
    lower_bound = median - 1 * std
    upper_bound = median + 1 * std
    if row['Hourly_Cost'] < lower_bound or row['Hourly_Cost'] > upper_bound:
        return median  # Replace outliers with group median
    return row['Hourly_Cost']  # Keep original value if not an outlier

# Apply the function to adjust the outliers
df['Hourly_Cost'] = df.apply(replace_outliers_with_median, axis=1)

# Drop unnecessary columns
df = df.drop(['mean', 'std'], axis=1)

# Verify the adjusted dataset
print(df[['Client_Country', 'EX_level_demand', 'Category', 'Hourly_Cost']].head())
```

```python
# Drop rows where 'Client_Country' is null
df = df.dropna(subset=['Client_Country'])


# Separate hourly and fixed datasets
hourly_features = df[df['Payment_type_Hourly'] == 1]
fixed_features = df[df['Payment_type_Hourly'] == 0]
hourly_target = df[df['Payment_type_Hourly'] == 1]['Hourly_Cost']
fixed_target = df[df['Payment_type_Hourly'] == 0]['Fixed_Cost']

```

```python
# Drop unnecessary columns
fixed_features = fixed_features.drop(columns=['Job_Cost', 'Hourly_Rate','Hourly_Cost', 'Fixed_Cost','Payment_type_Hourly'])

hourly_features = hourly_features.drop(columns=['Job_Cost', 'Hourly_Rate','Hourly_Cost', 'Fixed_Cost','Payment_type_Hourly'])
```

```python
fixed_features = pd.get_dummies(fixed_features, columns=['Category', 'Client_Country'], drop_first=True)
hourly_features = pd.get_dummies(hourly_features, columns=['Category', 'Client_Country'], drop_first=True)
hourly_features
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Number of splits
n_splits = 3

# Initialize K-Fold for hourly data
kf_hourly = KFold(n_splits=n_splits, shuffle=True, random_state=42)

# Initialize K-Fold for fixed data
kf_fixed = KFold(n_splits=n_splits, shuffle=True, random_state=42)

# Placeholder for metrics
hourly_mse_scores = []
hourly_r2_scores = []

fixed_mse_scores = []
fixed_r2_scores = []

# Hourly Data Cross-Validation
for fold, (train_idx, test_idx) in enumerate(kf_hourly.split(hourly_features)):
    X_hourly_train, X_hourly_test = hourly_features.iloc[train_idx], hourly_features.iloc[test_idx]
    y_hourly_train, y_hourly_test = hourly_target.iloc[train_idx], hourly_target.iloc[test_idx]
    
    # Train model (example using XGBoost)
    model = XGBRegressor(random_state=42,enable_categorical=True)
    model.fit(X_hourly_train, y_hourly_train)
    
    # Predict and evaluate
    y_hourly_pred = model.predict(X_hourly_test)
    mse = mean_squared_error(y_hourly_test, y_hourly_pred)
    r2 = r2_score(y_hourly_test, y_hourly_pred)
    
    hourly_mse_scores.append(mse)
    hourly_r2_scores.append(r2)
    
    print(f"Fold {fold + 1}: Hourly - MSE: {mse:.2f}, R²: {r2:.2f}")

# Fixed Data Cross-Validation
for fold, (train_idx, test_idx) in enumerate(kf_fixed.split(fixed_features)):
    X_fixed_train, X_fixed_test = fixed_features.iloc[train_idx], fixed_features.iloc[test_idx]
    y_fixed_train, y_fixed_test = fixed_target.iloc[train_idx], fixed_target.iloc[test_idx]
    
    # Train model (example using XGBoost)
    model = XGBRegressor(random_state=42,enable_categorical=True)
    model.fit(X_fixed_train, y_fixed_train)
    
    # Predict and evaluate
    y_fixed_pred = model.predict(X_fixed_test)
    mse = mean_squared_error(y_fixed_test, y_fixed_pred)
    r2 = r2_score(y_fixed_test, y_fixed_pred)
    
    fixed_mse_scores.append(mse)
    fixed_r2_scores.append(r2)
    
    print(f"Fold {fold + 1}: Fixed - MSE: {mse:.2f}, R²: {r2:.2f}")

# Average Metrics
print(f"\nHourly - Average MSE: {np.mean(hourly_mse_scores):.2f}, Average R²: {np.mean(hourly_r2_scores):.2f}")
print(f"Fixed - Average MSE: {np.mean(fixed_mse_scores):.2f}, Average R²: {np.mean(fixed_r2_scores):.2f}")
```

```python
# Split hourly data
X_hourly_train, X_hourly_test, y_hourly_train, y_hourly_test = train_test_split(
    hourly_features, hourly_target, test_size=0.3, random_state=42
)

# Split fixed data
X_fixed_train, X_fixed_test, y_fixed_train, y_fixed_test = train_test_split(
    fixed_features, fixed_target, test_size=0.3, random_state=42
)

```

```python
# Predictions
hourly_preds = hourly_model.predict(X_hourly_test)
fixed_preds = fixed_model.predict(X_fixed_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, median_absolute_error, r2_score
import numpy as np

def evaluate(y_test, predictions, normalization='range'):
    mse = mean_squared_error(y_test, predictions)
    mae = mean_absolute_error(y_test, predictions)
    medae = median_absolute_error(y_test, predictions)  # Calculate MedAE
    r2 = r2_score(y_test, predictions)
    rmse = np.sqrt(mse)  # Calculate RMSE
    
    if normalization == 'range':
        # NRMSE using range
        nrmse = rmse / (y_test.max() - y_test.min())
    elif normalization == 'mean':
        # NRMSE using mean
        nrmse = rmse / y_test.mean()
    else:
        raise ValueError("Invalid normalization method. Use 'range' or 'mean'.")
    
    return mse, mae, medae, r2, rmse, nrmse

# Evaluate hourly model
hourly_metrics = evaluate(y_hourly_test, hourly_preds, normalization='range')
print(f"Hourly Model - MSE: {hourly_metrics[0]}, MAE: {hourly_metrics[1]}, MedAE: {hourly_metrics[2]}, R2: {hourly_metrics[3]}, RMSE: {hourly_metrics[4]}, NRMSE: {hourly_metrics[5]}")

# Evaluate fixed model
fixed_metrics = evaluate(y_fixed_test, fixed_preds, normalization='range')
print(f"Fixed Model - MSE: {fixed_metrics[0]}, MAE: {fixed_metrics[1]}, MedAE: {fixed_metrics[2]}, R2: {fixed_metrics[3]}, RMSE: {fixed_metrics[4]}, NRMSE: {fixed_metrics[5]}")
```
