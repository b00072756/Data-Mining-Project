```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from catboost import CatBoostRegressor, Pool
import matplotlib.pyplot as plt
import shap
import lime.lime_tabular
import tqdm as notebook_tqdm
```

```python
df = pd.read_csv('ds_word2vec.csv')

# Create target variables based on Payment_type_Hourly
df['Hourly_Cost'] = np.where(df['Payment_type_Hourly'] == 1, df['Hourly_Rate'], np.nan)
df['Fixed_Cost'] = np.where(df['Payment_type_Hourly'] == 0, df['Job_Cost'], np.nan)

df['Category'] = df['Category'].astype('category')
df['Client_Country'] = df['Client_Country'].astype('category')

# Apply filters separately for hourly and fixed costs
df = df[(df['Payment_type_Hourly'] == 1) & (df['Hourly_Cost'] >= 0) & (df['Hourly_Cost'] <= 150) |
        (df['Payment_type_Hourly'] == 0) & (df['Fixed_Cost'] >= 0) & (df['Fixed_Cost'] <= 5000)]
df = df.dropna(subset=['Client_Country'])
```

```python
# Compute group statistics
group_stats = df.groupby(['Client_Country', 'Category'])['Fixed_Cost'].agg(['mean', 'std']).reset_index()

# Merge the stats back to the dataframe
df = df.merge(group_stats, on=['Client_Country', 'Category'], how='left')

# Function to replace outliers with group median
def replace_outliers_with_median(row):
    median = row['mean']  # Use median instead of mean
    std = row['std']
    lower_bound = median - 2 * std
    upper_bound = median + 2 * std
    if row['Fixed_Cost'] < lower_bound or row['Fixed_Cost'] > upper_bound:
        return median  # Replace outliers with group median
    return row['Fixed_Cost']  # Keep original value if not an outlier

# Apply the function to adjust the outliers
df['Fixed_Cost'] = df.apply(replace_outliers_with_median, axis=1)

# Drop unnecessary columns
df = df.drop(['mean', 'std'], axis=1)

# Verify the adjusted dataset
print(df[['Client_Country', 'EX_level_demand', 'Category', 'Fixed_Cost']].head())
```

```python
# Compute group statistics
group_stats = df.groupby(['Client_Country','Category'])['Hourly_Cost'].agg(['mean', 'std']).reset_index()

# Merge the stats back to the dataframe
df = df.merge(group_stats, on=['Client_Country','Category'], how='left')

# Function to replace outliers with group median
def replace_outliers_with_median(row):
    median = row['mean']
    std = row['std']
    lower_bound = median - 1 * std
    upper_bound = median + 1 * std
    if row['Hourly_Cost'] < lower_bound or row['Hourly_Cost'] > upper_bound:
        return median  # Replace outliers with group median
    return row['Hourly_Cost']  # Keep original value if not an outlier

# Apply the function to adjust the outliers
df['Hourly_Cost'] = df.apply(replace_outliers_with_median, axis=1)

# Drop unnecessary columns
df = df.drop(['mean', 'std'], axis=1)

# Verify the adjusted dataset
print(df[['Client_Country', 'EX_level_demand', 'Category', 'Hourly_Cost']].head())
```

```python
# Drop rows where 'Client_Country' is null
df = df.dropna(subset=['Client_Country'])


# Separate hourly and fixed datasets
hourly_features = df[df['Payment_type_Hourly'] == 1]
fixed_features = df[df['Payment_type_Hourly'] == 0]
hourly_target = df[df['Payment_type_Hourly'] == 1]['Hourly_Cost']
fixed_target = df[df['Payment_type_Hourly'] == 0]['Fixed_Cost']

```

```python
# Drop unnecessary columns
fixed_features = fixed_features.drop(columns=['Job_Cost', 'Hourly_Rate','Hourly_Cost', 'Fixed_Cost','Payment_type_Hourly'])

hourly_features = hourly_features.drop(columns=['Job_Cost', 'Hourly_Rate','Hourly_Cost', 'Fixed_Cost','Payment_type_Hourly'])
```

```python
fixed_features = pd.get_dummies(fixed_features, columns=['Category', 'Client_Country'], drop_first=True)
hourly_features = pd.get_dummies(hourly_features, columns=['Category', 'Client_Country'], drop_first=True)
```

```python
# Split hourly data
X_hourly_train, X_hourly_test, y_hourly_train, y_hourly_test = train_test_split(
    hourly_features, hourly_target, test_size=0.2, random_state=42
)

# Split fixed data
X_fixed_train, X_fixed_test, y_fixed_train, y_fixed_test = train_test_split(
    fixed_features, fixed_target, test_size=0.2, random_state=42
)
```

```python
# Initialize LightGBM models
hourly_model = LGBMRegressor(n_estimators=500, learning_rate=0.1, max_depth=6, random_state=42)
fixed_model = LGBMRegressor(n_estimators=500, learning_rate=0.1, max_depth=6, random_state=42)


# Clean feature names for all datasets
def clean_column_names(df):
    df.columns = df.columns.str.replace(r"[^a-zA-Z0-9_]", "_", regex=True)
    return df

# Apply to all feature sets
X_hourly_train = clean_column_names(X_hourly_train)
X_hourly_test = clean_column_names(X_hourly_test)
X_fixed_train = clean_column_names(X_fixed_train)
X_fixed_test = clean_column_names(X_fixed_test)

```

```python
# Ensure feature names are unique by appending a suffix to duplicates
def make_unique_columns(df):
    seen = set()
    new_columns = []
    for col in df.columns:
        new_col = col
        i = 1
        while new_col in seen:
            new_col = f"{col}_{i}"
            i += 1
        seen.add(new_col)
        new_columns.append(new_col)
    df.columns = new_columns
    return df

# Apply to the feature sets
X_hourly_train = make_unique_columns(X_hourly_train)
X_hourly_test = make_unique_columns(X_hourly_test)
X_fixed_train = make_unique_columns(X_fixed_train)
X_fixed_test = make_unique_columns(X_fixed_test)
```

```python
# Train models
hourly_model.fit(X_hourly_train, y_hourly_train)
fixed_model.fit(X_fixed_train, y_fixed_train)
```

```python
# Feature Importance
hourly_importance = pd.DataFrame({
    'Feature': X_hourly_train.columns,
    'Importance': hourly_model.feature_importances_
}).sort_values(by='Importance', ascending=False)
print("Hourly Model Feature Importance:")
print(hourly_importance)
```

```python
# Feature Importance
hourly_importance = pd.DataFrame({
    'Feature': X_hourly_train.columns,
    'Importance': hourly_model.feature_importances_
}).sort_values(by='Importance', ascending=False)
print("Hourly Model Feature Importance:")
print(hourly_importance)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, median_absolute_error, r2_score
import numpy as np
# Predictions
hourly_preds = hourly_model.predict(X_hourly_test)
fixed_preds = fixed_model.predict(X_fixed_test)
def evaluate(y_test, predictions, normalization='range'):
    mse = mean_squared_error(y_test, predictions)
    mae = mean_absolute_error(y_test, predictions)
    medae = median_absolute_error(y_test, predictions)  # Calculate MedAE
    r2 = r2_score(y_test, predictions)
    rmse = np.sqrt(mse)  # Calculate RMSE
    
    if normalization == 'range':
        # NRMSE using range
        nrmse = rmse / (y_test.max() - y_test.min())
    elif normalization == 'mean':
        # NRMSE using mean
        nrmse = rmse / y_test.mean()
    else:
        raise ValueError("Invalid normalization method. Use 'range' or 'mean'.")
    
    return mse, mae, medae, r2, rmse, nrmse

# Evaluate hourly model
hourly_metrics = evaluate(y_hourly_test, hourly_preds, normalization='range')
print(f"Hourly Model - MSE: {hourly_metrics[0]}, MAE: {hourly_metrics[1]}, MedAE: {hourly_metrics[2]}, R2: {hourly_metrics[3]}, RMSE: {hourly_metrics[4]}, NRMSE: {hourly_metrics[5]}")

# Evaluate fixed model
fixed_metrics = evaluate(y_fixed_test, fixed_preds, normalization='range')
print(f"Fixed Model - MSE: {fixed_metrics[0]}, MAE: {fixed_metrics[1]}, MedAE: {fixed_metrics[2]}, R2: {fixed_metrics[3]}, RMSE: {fixed_metrics[4]}, NRMSE: {fixed_metrics[5]}")
```
